# Veri Okuryazarlığı

## 1. Merkezi Eğilim Ölçüleri

### 1.1. Aritmetik Ortalama
Bir seride (değişkende) yer alan tüm değerlerin toplamının değerler sayısına bölünmesi ile elde edilir.



Seri dağılımı simetrik ise aritmetik ortalama tercih edilebilir

[5,7,8,12,25,18,20]

### 1.2. Median (Medyan)
Bir seri küçükten büyüğe sıralandığında serinin orta noktasındaki değere medyan denir.



Seri dağılımı simetrik değil ise median tercih edilebilir.

[5,7,8,12,25,18,200]

### 1.3. Mode (Mod)
Bir seride en çok tekrar eden değer Mode değeridir.


[5,7,8,12,12,12,25,18,200]      Mode=12

### 1.4. Quartiles 
Küçükten büyüğe sıralanan bir seriyi çeyrekliklere ayıran değerlere denir.






[5,7,8,12,25,18,200]
Q1=7
Q2=12
Q3=18

<b>Not:</b> iki emlakçı örneği; 


A örneğinde ortalama ve medyan değerleri yakın olduğundan veriseti homojen dağılmıştır yorumu yapılabilir ve ikisi de kullanılabilir.

B örneğinde ortalama ve medyan birbirlerine uzak değerlerdir ve verisetinin homojen dağılmadığının bir göstergesidir. Bu durumda medyan tercih edilmelidir.

## 2. Dağılım Ölçüleri (Measure of Dispersion)

### 2.1 Range (Değişim Aralığı)
Serideki maksimum değerden minimum değerin çıkartılmasıyla elde edilir.

Range=max-min


### 2.2. Standard Deviation (Standart Sapma)
Standart sapma veri değerlerinin aritmetik ortalamadan farklarının karelerinin toplamının veri sayısı -1'e bölümünün kareköküdür, yani verilerin ortalamadan sapmalarının kareler ortalamasının karekökü olarak tanımlanır.




### 2.3. Variance (Varyans)
Standart sapmanın karesidir.




### 2.4. Skewness (Çarpıklık)
Bir değişkenin dağılımının simitrik olmama durumudur.


Pearson çarpıklık katsayısı:



[-1,1] aralığında bir sonuç verecektir. 

PÇK < 0 == Negatif çarpıklık

PÇK > 0 == Pozitif çarpıklık

PÇK = 0 == simetrik

### 2.5. Kurtosis (Basıklık)
Bir değişkenin dağılımının basıklığını gösterir.



üç farkı grubun yaş dağılımına baktığımızı düşünelim. dik çizgi 30 yaşı ifade ettiğini varsayalım.

Leptokurtic == gruptakilerin 30 yaşa daha yakın olan bireylerle dolu olduğunu

Mesokurtic  == normal dağılıma uygun olduğunu

platykurtic == basık bir dağılım oduğunu görüyoruz ve gruptakilerin yaşlarının değişken aralığında yayvan bir şekilde dağıldığını anlıyoruz. standart sapmanın yüksek çıkacağını düşünebiliriz.

#### 3.1.4. Verinin Analiz Edilmesi ve Yorumlanması

Gündüz öğrencilerinde ise bu durumun tam tersi olduğu ve notların grup içerisinde istikrarlı bir şekilde dağıldığı yorumu yapılabilir.
