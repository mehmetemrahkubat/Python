### 2.2. Gözlem ve Değişken Seçimi: loc ve iloc

```python
import numpy as np
import pandas as pd
a = np.random.randint(1,30, size = (10,3))
df = pd.DataFrame(a, columns = ["var1","var2","var3"])
df
```

**loc:** is label-based, which means that you have to specify rows and columns based on their row and column labels.
**iloc:** is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

```python
print(df.loc[0:3]) #3 dahildir
print(df.iloc[0:3]) #3 dahil değildir
print(df.iloc[0,0])
```

```python
print(df.iloc[:3,:2])
```

### 2.7. İleri Aggregation, Filtering, Transform ve Apply işlemleri
**Toplulaştırma**

```python
import pandas as pd
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate([min, np.median, max]) 
```

```python
df.groupby("gruplar").aggregate({"degisken1": min, "degisken2": max})
```

**filter**

```python
import pandas as pd
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

```python
def filter_func(x):
    return x["degisken1"].std() > 9

print(df.groupby("gruplar").std())

print(df.groupby("gruplar").filter(filter_func))
```

**transform**

```python
import pandas as pd
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

```python
df["degisken1"]*9
```

```python
df_a = df.iloc[:,1:3] #grupları almıyoruz ki numerik degiskenlerde uygulayabilelim
df_a
```

```python
df_b=df_a.transform(lambda x: (x-x.mean()) / x.std())
df_b
```

**apply**

```python
import pandas as pd
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

## 3. Pivot Tablolar
Veriseti üzerinde bazı satır sütun işlemleri ile kullanıma uygun tablolar hazırlamaya yarar. Benzer işlemler groupby ile de yapilabilir.

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby("sex")["survived"].mean()
```

```python
titanic.groupby("sex")[["survived"]].mean() #iki koseli parantez bunu bir dataframe'e cevirir
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean")
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack() #unstack() hiyerarşik index yapisini cozer
```

```python
#pivot table
titanic.pivot_table("survived", index = "sex", columns = "class")
```

```python
# yaş sürekli değişkenini belirli aralıklar alarak kategorik değişkene çevirip oluşan sınıfları da pivot tabloya değişken olarak ekleyelim
print(titanic.age.head())
age=pd.cut(titanic["age"],[0,20,40,60,80])
print(age.head(10))
```

```python
titanic.pivot_table("survived", ["sex", age], "class")
```

```python
print(titanic.groupby(["sex","class"]).count())
print(titanic.groupby(["sex","class"])["age"].count())
```

```python
print(titanic.groupby(["sex","class"])["age"].count())
print(titanic.groupby(["sex","class",titanic.survived==1])[["age"]].count().unstack())
```

```python
print(titanic.groupby(["sex","class",age])["age"].count())
print(titanic.groupby(["sex","class",age,titanic.survived==1])[["age"]].count().unstack())
```

```python
df1=titanic.groupby(["sex","class",age])["age"].count()
df2=titanic.groupby(["sex","class",age,titanic.survived==1])["age"].count().unstack()
df3=pd.concat([df1,df2], axis=1)
df3
```
